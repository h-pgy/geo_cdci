```python
mapa = folium.Map(location=[-23.5505, -46.6333], zoom_start=14, tiles=None)

folium.WmsTileLayer(
    url="https://wms.geosampa.prefeitura.sp.gov.br/geoserver/geoportal/ows",
    layers="geoportal:MapaBase_Politico",
    fmt="image/png",
    transparent=True,
    version="1.1.1",
    #crs="EPSG:4326",
    attr="GeoSampa"
).add_to(mapa)

mapa.save("mapa_geosampa.html")
```
